```python
def calculate_object_height(ref_data, HEIGHT, data):
   ref_data = np.array(ref_data)[:, 8:34]
   MAX = np.max(np.amax(ref_data, axis=1))
   newData = np.array(data)[8:34]
   height_max = HEIGHT * np.max(newData) / MAX
   print('HEIGHT MAX=', "{:.2f}".format(height_max))
   return height_max
```

```python
distanceReferenceData = result.loc[result['distance']==80]
distanceReferenceData = distanceReferenceData.iloc[:, 5:].values
calculate_object_height(distanceReferenceData, 15.5, df_cobble2.values[3][4:])
```
